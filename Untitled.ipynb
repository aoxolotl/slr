```python

def show_signseq(utt, ann, figsize):
    """Shows the frames of an utterance and the corresponding annotation
    """
    total_frames = len(utt)
    plt.rcParams['figure.figsize'] = (figsize, figsize)
    for idx, imfile in enumerate(utt):
        img = Image.open(imfile)
        plt.subplot( 1, total_frames, idx + 1)
        imgplot = plt.imshow(img)
        plt.axis('off')
    plt.show()
    print(ann)
        
```

```python
google_net = torchvision.models.inception_v3(pretrained=True)
```

```python
class RWTHSignDataset(Dataset):
    """RWTH Phoenix dataset"""
    def __init__(self, csv_file):
        df = pd.read_csv(csv_file, delimiter='|')
        sign_folders = list(df['folder'])
        self.annotations = df['annotation']
        self.frame_files = [sorted(glob(os.path.join(sign_dir, f))) for f in sign_folders]
        
    def __len__(self):
        return len(self.annotations)
    
    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()
        frames = [Image.open(f) for f in self.frame_files[idx]]
        sample = {'frames': frames, 'annotations': self.annotations[idx]}
        
        return sample
```

```python
d = RWTHSignDataset("/home/reverie/datasets/phoenix2014-release/phoenix-2014-multisigner/annotations/manual/train.corpus.csv")
```

```python
preprocess = transforms.Compose([
    transforms.Resize(256),
    #transforms.CenterCrop(299), # inception_v3 uses 299x299 as initial image size
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
for t in d:
    try:
        img_t = preprocess(t['frames'])
        print(t['annotations'])

        input_batch = img_t

        # move to gpu
        if torch.cuda.is_available():
            input_batch = input_batch.to('cuda')
            google_net.to('cuda')

        google_net.eval()
        with torch.no_grad():
            output = google_net(input_batch)

        print(np.argmax(torch.nn.functional.softmax(output[0], dim=0).cpu(), -1))
    except:
        continue
```

```python
class Autoencoder(nn.Module):
    def __init__(self):
        super(Autoencoder, self).__init__()
        
        self.encoder = nn.Sequential(
            nn.Conv2d(3, 6, kernel_size=5),
            nn.ReLU(True),
            nn.Conv2d(6,16,kernel_size=5),
            nn.ReLU(True),
            nn.MaxPool2d(2, stride=1)
        )
        
        self.decoder = nn.Sequential(             
            nn.ConvTranspose2d(16,6,kernel_size=5),
            nn.ReLU(True),
            nn.ConvTranspose2d(6,3,kernel_size=5),
            nn.ReLU(True),
            nn.Sigmoid(),
            nn.ConvTranspose2d(8, 1, 2, stride=2, padding=1)
        )    
        
    def forward(self,x):
        x = self.encoder(x)
        x = self.decoder(x)
        return x
```

```python
class RWTHAllImg(Dataset):
    """Class to load all images from RWTH dataset"""
    def __init__(self, globpattern, transform=None):
        self.imglobs = globpattern
        self.transform = transform
    
    def __getitem__(self, idx):
        if torch.is_tensor(idx):
            idx = idx.tolist()
        img = Image.open(self.imglobs[idx])
        if self.transform:
            img = self.transform(img)
        sample = {'image': img}
        return sample
    
    def __len__(self):
        return len(self.imglobs)
```

```python
traindir = "/home/reverie/datasets/phoenix2014-release/phoenix-2014-multisigner/features/fullFrame-210x260px/train"
train_glob = glob(os.path.join(traindir, "*/*/*.png"))
print(len(train_glob))
train_glob = np.random.choice(train_glob, 10000, replace=False)
trainloader = torch.utils.data.DataLoader(RWTHAllImg(train_glob, transform=preprocess), batch_size=128)

testdir = "/home/reverie/datasets/phoenix2014-release/phoenix-2014-multisigner/features/fullFrame-210x260px/test"
test_glob = glob(os.path.join(testdir, "*/*/*.png"))
print(len(test_glob))
test_glob  = np.random.choice(test_glob, 1000, replace=False)
testloader = torch.utils.data.DataLoader(RWTHAllImg(test_glob, transform=preprocess), batch_size=128)
```

```python
model = Autoencoder().cuda()
```

```python
distance = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(),weight_decay=1e-3)
num_epochs = 10
```

```python
for epoch in range(num_epochs):
    itr = 1
    for img in trainloader:
        #img, _ = data
        img = img['image'].cuda()
        # ===================forward=====================
        output = model(img)
        loss = distance(output, img)
        # ===================backward====================
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if (itr % 50) == 0:
            print(f"Iteration {itr}...")
        itr += 1
    # ===================log========================
    print('epoch [{}/{}], loss:{:.4f}'.format(epoch+1, num_epochs, loss))
```
