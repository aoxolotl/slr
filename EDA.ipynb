```python
import os
import numpy as np
from matplotlib import pyplot as plt
import glob
```

```python
img_files = glob.glob('data/*.png')
sub_files = glob.glob('data/*.txt')
```

```python
word_seq = []
seq_len = []
for sub_file in sub_files:
    f = open(sub_file)
    line = f.read()
    line = line.strip().lower()
    match = os.path.splitext(sub_file)[0]
    img_files = glob.glob(f"{match}*.png")
    word_seq.append((line.split(), img_files))
    seq_len.append((len(line.split()), len(img_files)))
print(word_seq[9])
```
